```python
ds = Dataset("0b381b77-2506-4367-a675-9373e42e4a75")
images = ds.dataset_info()
```

```python
import matplotlib.pyplot as plt
import imageio
import os

def show_image(_class, path):
    plt.title("Class: \"%s\"    %s" % (_class, os.path.basename(path)) )
    plt.imshow(imageio.imread(path, as_gray=False, pilmode="RGB"))
    plt.show()

for _class, paths in images.items():
    for path in paths:
        show_image(_class, path)
```
